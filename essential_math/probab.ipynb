```python
from scipy.stats import beta
def calculate_beta_dist(alpha: int, beta_ : int, prob: float, side: str):
    """ Please include the side to calculate the probability distribution """
    b = beta.cdf(prob, alpha, beta_)
    if side == "left":
        return b
    else: 
        return 1 - b
calculate_beta_dist(137,50,.40, 'left')
```

```python
def calculate_binomial(success: int, n: int, prob: float, side: str) -> float:
    b  = binom.cdf(success, n, prob)
    if side == "left": 
        return b
    else: 
        return 1-b
calculate_binomial( 50, 137, .40, 'right')
```
