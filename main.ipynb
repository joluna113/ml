## Creating Scatter Plots
We need some scatter plots to see the relationship of two different variables


```python
cases = df['cases'].to_numpy()
pop = df['population'].to_numpy()
gdp = df['gdp'].to_numpy()
urban = df['urban'].to_numpy()
```

```python
N = len(cases)
plt.scatter(range(N), pop, label="Population")
plt.scatter(range(N), cases, label="Cases")
plt.title('COVID-19 cases through April 30,2020')
plt.xlabel("State Number")
plt.ylabel("Number of Cases")
plt.legend();
```

We cab see that the axis are little bit off since we are using 2 variables. 
There is a difference in scale. Population bigger compared to the number of cases. 



```python
fig, ax = plt.subplots()
ax.scatter(range(N), cases/1000, label="Cases")
ax.set_ylabel("Number of Cases")
ax2 = ax.twinx()
ax2.scatter(range(N), pop/1e6, color='C1' , label="Population", marker="*")
ax2.set_ylabel("Population")
ax.legend(loc="upper left")
ax2.legend(loc="upper right")
ax.spines['right'].set_visible(True);
```
